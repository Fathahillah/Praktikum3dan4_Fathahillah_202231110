```python
import numpy as np
from skimage import io, color
import matplotlib.pyplot as plt
```

-numpy: Digunakan untuk manipulasi array dan operasi matematika lainnya

-skimage.io: Digunakan untuk membaca dan menulis gambar

-skimage.color: Digunakan untuk konversi ruang warna

-matplotlib.pyplot: Digunakan untuk menampilkan gambar

```python
image = io.imread('2.jpg')
```

io.imread: Fungsi ini digunakan untuk memuat gambar dari jalur file yang diberikan. Gambar akan dimuat sebagai array NumPy

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
```

-plt.figure: Membuat sebuah figure baru dengan ukuran yang ditentukan

-plt.subplot: Membuat sub-plot di dalam figure, dalam hal ini ada 1 baris dan 2 kolom, dan kita memilih sub-plot pertama

-plt.imshow: Menampilkan gambar

-plt.title: Menambahkan judul pada sub-plot

-plt.axis('off'): Menyembunyikan sumbu koordinat

```python
hsv_image = color.rgb2hsv(image)
```

color.rgb2hsv: Fungsi ini mengkonversi gambar dari ruang warna RGB ke ruang warna HSV. Gambar hasil konversi disimpan dalam variabel hsv_image

```python
plt.subplot(1, 2, 2)
plt.imshow(hsv_image)
plt.title('HSV Image')
plt.axis('off')
plt.show()
```

-plt.subplot(1, 2, 2): Memilih sub-plot kedua

-plt.imshow: Menampilkan gambar dalam ruang warna HSV

-plt.show: Menampilkan semua figure dan sub-plot yang telah didefinisikan

```python
hue, saturation, value = hsv_image[:, :, 0], hsv_image[:, :, 1], hsv_image[:, :, 2]

plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.imshow(hue, cmap='hsv')
plt.title('Hue')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(saturation, cmap='gray')
plt.title('Saturation')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(value, cmap='gray')
plt.title('Value')
plt.axis('off')

plt.show()
```

-hue, saturation, value: Memisahkan komponen Hue, Saturation, dan Value dari gambar HSV. hsv_image[:, :, 0] adalah komponen Hue, hsv_image[:, :, 1] adalah komponen Saturation, dan hsv_image[:, :, 2] adalah komponen Value.
plt.subplot(1, 3, 1), plt.subplot(1, 3, 2), plt.subplot(1, 3, 3): Membuat 3 sub-plot untuk masing-masing komponen

-cmap='hsv': Menggunakan colormap HSV untuk menampilkan komponen Hue

-cmap='gray': Menggunakan colormap grayscale untuk menampilkan komponen Saturation dan Value

-plt.show: Menampilkan semua figure dan sub-plot yang telah didefinisikan
